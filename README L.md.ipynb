
Project Title:
Library Management System 

Project Description:
This project is a menu-driven Library Management System developed using Python.
It helps in managing library books by allowing the user to add books, view all
books, search for a book using Book ID, and save the data into a CSV file.

The program uses Python concepts such as lists, dictionaries, functions, loops,
conditional statements, exception handling, and file handling.

Features:
1. Add new book details (Book ID, Title, Author, Copies)
2. View all available books
3. Search for a book using Book ID
4. Save library data to a CSV file
5. Menu-driven program
6. Handles invalid input using exception handling

Concepts Used:
- Lists
- Dictionaries
- Functions
- While loop
- If / Elif / Else
- Exception handling (try-except)
- File handling (CSV)
- Pandas library
- String methods (strip)

How to Run the Program:
1. Open Jupyter Notebook or Python IDE
2. Run the program file
3. The menu will be displayed
4. Enter a choice from 1 to 5 to perform operations

Menu Options:
1. Add Book
2. View Books
3. Search Book
4. Save to File
5. Exit

Input Guidelines:
- Book ID must be an integer
- Number of copies must be an integer
- Enter valid inputs to avoid errors

File Handling Details:
- Library data is saved in a file named 'library.csv'
- The file is created using pandas DataFrame
- File is saved in the same directory as the program

Data Structure Used:
- Dictionary with lists
  Keys used:
  - Book_ID
  - Title
  - Author
  - Copies
