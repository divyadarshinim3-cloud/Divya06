```python
import pandas as pd
library = {
    "Book_ID":[101,102,103,104,105],
    "Title":["Python basics","Data science","AI concepts","DBMS","Operating systems"],
    "Author":["Guido","James","Andrew","Elmasri","Silberschatz"],
    "Copies":[5,3,4,6,2]
}
library             
```

```python
df = pd.DataFrame(library)
df
```

```python
def add_book():
    try:
        book_id = int(input("Enter Book ID: "))
        title = input("Enter Book Title: ").strip()
        author = input("Enter Author Name: ").strip()
        copies = int(input("Enter Number of Copies: "))

        library["Book_ID"].append(book_id)
        library["Title"].append(title)
        library["Author"].append(author)
        library["Copies"].append(copies)

        print("Book added successfully!")

    except ValueError:
        print("Invalid input! Please enter correct values.")
```

```python
def view_books():
    if len(library["Book_ID"]) == 0:
        print("No books available.")
    else:
        df = pd.DataFrame(library)
        print(df)
```

```python
def search_book():
    try:
        book_id = int(input("Enter Book ID to search: "))
        df = pd.DataFrame(library)

        result = df[df["Book_ID"] == book_id]

        if result.empty:
            print("Book not found.")
        else:
            print("Book Found:")
            print(result)

    except ValueError:
        print("Invalid input!")
```

```python
def save_to_file():
    df = pd.DataFrame(library)
    df.to_csv("library.csv", index=False)
    print("Library data saved to library.csv")
```

```python
# while True:
    print("\n--- Library Management System ---")
    print("1. Add Book")
    print("2. View Books")
    print("3. Search Book")
    print("4. Save to File")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_book()
    elif choice == "2":
        view_books()
    elif choice == "3":
        search_book()
    elif choice == "4":
        save_to_file()
    elif choice == "5":
        print("Thank you!")
        break
    else:
        print("Invalid choice! Try again.")
```
